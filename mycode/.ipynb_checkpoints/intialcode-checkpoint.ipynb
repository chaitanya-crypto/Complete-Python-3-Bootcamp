This below opreation performs all operators of calc Numeric operations

`+`, `-`, `*`, `/`, `%`
and `power` is `**`
and for `Floor Division` we use `//`

```python
print("Add : ", 5+3)
print("Sub : ", 8-6)
print("Mul : ", 3*2)
print("Div : ", 10/5)
print("Mod : ", 10%3)
print("Power : ", 2**3)
print("Floor Division : ", 10//3)
```
