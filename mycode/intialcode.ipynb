This section cover String manpulation splicing with strings

```python
mystring = 'hello world'
```

```python
print('mystring[0] : ', mystring[0])
print('reverse indexing by using -1 : ', mystring[-1])
print('mystring[0:3] it starts from zero and goes till 3 index and doesnt include 3 index : ', mystring[0:3])
print('mystring[3:] it starts from 3 index and goes till end : ', mystring[3:])
print('mystring[:3] it starts from 0 index and goes till 3 index and doesnt include 3 index : ', mystring[:3])
print('mystring[::1] it starts from 0 index and goes till end by taking 1 step : ', mystring[::1])
print('mystring[::2] it starts from 0 index and goes till end by taking 2 step : ', mystring[::2])
print('mystring[2:9] it starts from 2 index and goes till 9 index and doesnt include 9 index : ', mystring[2:9])
print('mystring[2:9:2] it starts from 2 index and goes till 9 index and doesnt include 9 index by taking 2 step : ', mystring[2:9:2])
print('mystring[::-1] it starts from end and goes till 0 index by taking -1 step : ', mystring[::-1]) # this is used to reverse a string
```

`Loops` in pyhton

```python
my_list = [1,'sdsd',[1,2],{"key": "hello"},5,63,5,25,2,6,(1,[856]),21,3,7,468,4,656,None]

for num in my_list:
    if(type(num) is int and num % 2  == 0):
        print(f"Even Number is : {num}")
    elif(type(num) is int and num % 2  != 0):
        print(f'Odd Number is : {num}')
    elif(type(num) is str):
        print('skipping found string')
    elif(type(num) is dict):
        print('skiping found a dict')
    elif(type(num) is list):
        print('skiping found a list under list')
    elif(type(num) is tuple):
        print('skipping found a tuple')
    else:
        print('found NONE')
```
